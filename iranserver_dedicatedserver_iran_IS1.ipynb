```python
try:
    import urllib.request as urllib2
except ImportError:
    import urllib2

url= 'http://www.iranserver.com/vps/iran/'
response = urllib2.urlopen(url)
content = response.read()
# lines = content.decode("utf8").split("\n")

# for temp in lines:
#     print (temp)

```

```python
import pandas as pd

# dic = "https://docs.google.com/spreadsheets/d/18nWiecKkH7Oj5c2yPrqZPq5cFyieJeowVyCK107ZOYM/pub?gid=0&single=true&output=csv"
# data = pd.read_csv(dic)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(content, 'html.parser')
table = soup.find("table", {"class" : "panels tahoma"})
table2 = soup.find("table", {"class" : "tahoma tbl"})
        
count = len(table.findAll('tr')[0].findAll('td'))
records = []
for tr in table.findAll("tr"):
    trs = tr.findAll("td")
    record = []
    for i in range (count):
        record.append(trs[i].text.strip())
    
    records.append(record)

    
for tr in table2.findAll("tr"):
    if tr.find(class_='titr'):
        continue
    trs = tr.findAll("td")
    record = []
    for i in range (count):
        if(trs[i].text!=""):
            record.append(trs[i].text.strip())
        else:
            record.append(trs[i].span['class'])
        
    records.append(record)

df = pd.DataFrame(data=records)

df.set_value(1,0,"CPUTYPE")

strs_ = ["VPS" for x in range(len(df.columns)-1)]
strs2_ = [url for x in range(len(df.columns)-1)]
strs3_ = ["1" for x in range(len(df.columns)-1)]

strs = ['SERVICETYPE']
strs2 = ['PROVIDERLINK']
strs3 = ['PROVIDERID']

strs.extend(strs_)
strs2.extend(strs2_)
strs3.extend(strs3_)

# strs
df2 = pd.DataFrame([strs], columns=[0,1,2,3,4])
df3 = pd.DataFrame([strs2], columns=[0,1,2,3,4])
df4 = pd.DataFrame([strs3], columns=[0,1,2,3,4])

df = df.append(df2, ignore_index=True)
df = df.append(df3, ignore_index=True)
df = df.append(df4, ignore_index=True)


df[0] = df[0].str.replace(u'\u200c', ' ')
df
```
