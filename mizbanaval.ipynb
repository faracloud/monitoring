```python

import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
from collections import OrderedDict

page = requests.get("http://mizbanaval.com/vps/")
# print (page.content)

soup = BeautifulSoup(page.content, 'html.parser')
# print(soup.prettify())

print("*************************************************")
# print(list(soup.children))

wrap=(soup.find_all('div', class_='gw-go-col-wrap gw-go-col-wrap-0 gw-go-hover gw-go-disable-enlarge gw-go-disable-hover'))
wrap= wrap+soup.find_all('div', class_='gw-go-col-wrap gw-go-col-wrap-1 gw-go-hover gw-go-disable-enlarge gw-go-disable-hover')
wrap= wrap+soup.find_all('div', class_='gw-go-col-wrap gw-go-col-wrap-2 gw-go-hover gw-go-disable-enlarge gw-go-disable-hover')
wrap= wrap+soup.find_all('div', class_='gw-go-col-wrap gw-go-col-wrap-3 gw-go-hover gw-go-disable-enlarge gw-go-disable-hover')
link=soup.find_all('a', class_='gw-go-btn gw-go-btn-medium yp_onscreen')
print(link)

vpsdata=[]
i=1
for item1 in wrap:
    print("{}".format(i))
    i+=1
#     print(item1)
    item1= "<html> <head> <title>A simple example page</title> </head> <body>{}</body>".format(item1)
    soup2 = BeautifulSoup(item1, 'html.parser')
    vpsList =  ["SERVICETYPE : VPS"]
    vpsList =  vpsList + ["PROVIDERID : 2"]
    vpsList = vpsList + ["SERVICENAME :{}".format(pt.get_text()) for pt in soup2.select(".gw-go-header-top h3")]

    vpsList=vpsList+[t.get_text() for t in soup2.find_all('div', class_="gw-go-body-cell")]
    vpsList.append("INSTALLATION : 0")
    vpsList.append("OPERATION : {}".format(soup2.find(class_="gw-go-coinb").get_text()))
    vpsList.append("PROVIDERLINK : {}".format(soup2.find('a', class_="gw-go-btn gw-go-btn-medium",href=True)['href']))
    # vpsList.append("Name :{}".format(soup2.find(class_="gw-go-header-top").get_text()))
    vpsList = vpsList + ["OS : Linux/Windows"]
    vpsList = vpsList + ["DC : DE"]
    vpsList = vpsList + ["SUPPORT : سطح 1"]
    vpsList = vpsList + ["DELIVERY : 30min upto 4h "]
    vpsList = vpsList + ["VIRTUALIZATION : VMware ESX "]
    vpsList = vpsList + ["PANEL : انتخابی"]
    #حتما اضافه شود
#     vpsList = vpsList + ["CPUTYPE : {}".format()]
    # print( [([item2 for item2 in item.split(" : ")]) for item in vpsList])
    vpsDict=OrderedDict([[item2.strip() for item2 in item.split(" : ")] for item in vpsList])
    vpsDict['BW'] = vpsDict.pop('Connection')
    vpsDict['DISK'] = vpsDict.pop('Hard Disk')
    vpsDict['TRAFFIC'] = vpsDict.pop('Traffic')
    vpsDict['RAM'] = vpsDict.pop('Ram')
    vpsDict['DISKTYPE']="HDD" if "HDD" in vpsDict['DISK'] else "SSD"
    vpsDict['DISK']=vpsDict['DISK'].replace('SSD','')
    vpsDict['DISK']=vpsDict['DISK'].replace('HDD','')
    vpsDict['OPERATION'] = vpsDict['OPERATION'].replace('تومان','')
    vpsDict['OPERATION'] = vpsDict['OPERATION'].replace('.', '')
   

    print(vpsDict)

    vpsdata.append(vpsDict)
    # print(vpsList)

# print(vpsdata)

```
