```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import unicodecsv
from collections import OrderedDict

page = requests.get("https://asemanhost.com/vps-linux")
# print (page.content)

soup = BeautifulSoup(page.content, 'html.parser')
# print(soup.prettify())

```

```python

Header= soup.find_all('ul', class_='pricing_column_first')
detailRow=(soup.find_all('ul', class_='pricing_column gradient_red'))

Header
```

```python
vpsdata=[]
for item1 in detailRow:
    
    item1 = "<html> <head> <title>A simple example page</title> </head> <body>{}</body>".format(item1)
    soup2 = BeautifulSoup(item1, 'html.parser')
    # htmlHeader = "<html> <head> <title>A simple example page</title> </head> <body>{}</body>".format(Header[0])
    # Headersoup = BeautifulSoup(htmlHeader, 'html.parser')
    # HeaderItemList = [t.get_text() for t in Headersoup.find_all('li', class_="valuetitle")]
    # print(HeaderItemList)
    
    vpsDICT = {t["data-table"] : (t.get_text() if t.get_text()!='' else "yes" if t["class"]=="pricing_yes" else "no" )  for t in soup2.find_all('li',{"data-table":True})}
    vpsDICT["OS"]= "Linux"
    vpsDICT["DC"]="DE/US"             
    vpsDICT["VIRTUALIZATION"]= "VMware ESX "
    vpsDICT["PROVIDERID"] ="3"
    vpsdata.append(vpsDICT)

```

```python
df=pd.DataFrame(vpsdata)
df
```

```python
df.to_csv("asman.csv")
```
