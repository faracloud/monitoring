```python
try:
    import urllib.request as urllib2
except ImportError:
    import urllib2

response = urllib2.urlopen('http://hostiran.net/')
content = response.read()
# lines = content.decode("utf8").split("\n")

# for temp in lines:
#     print (temp)

```

```python
import pandas as pd

# dic = "https://docs.google.com/spreadsheets/d/18nWiecKkH7Oj5c2yPrqZPq5cFyieJeowVyCK107ZOYM/pub?gid=0&single=true&output=csv"
# data = pd.read_csv(dic)

from bs4 import BeautifulSoup

soup = BeautifulSoup(content, 'html.parser')
table = soup.find("table", {"class" : "responsive host table table-curved"})
# table2 = soup.find("table", {"class" : "tahoma tbl"})
table
```

```python
count = len(table.findAll('tr')[0].findAll('th'))

records = []

for tr in table.findAll("tr"):
    trs = tr.findAll("th")
    record = []
    if (len(trs)>0):
        for i in range (count):
            record.append(trs[i].text)

        records.append(record)

for tr in table.findAll("tr"):
    trs = tr.findAll("td")
    record = []
    if (len(trs)>0):
        for i in range (count):
            if(trs[i].text!=""):
                record.append(trs[i].text)
            else:
                record.append(trs[i].i['class'][1])
        records.append(record)


df = pd.DataFrame(data=records)
df

    
```

```python
df.T.to_csv("HI1.csv")
# df.T.to_csv("data.csv",sep='\t', encoding='utf-8')
```
